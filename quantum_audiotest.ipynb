```python
import librosa
import quantumaudio
from IPython.display import Audio, display

def play(array, rate=44100, autoplay=False):
  audio = Audio(data=array, rate=rate, autoplay=autoplay)
  display(audio)
  
y, sr = librosa.load('pianolick.mp3', sr = 22050, mono = True)
print(f'samples: {y.shape}, Sample Rate: {sr}')
play(y, sr)
```

```python
processed_data_qpam = quantumaudio.stream(y, shots=8000) # processing with QPAM (quantum probability amplitude modulation, faster (<1m for 5s mp3 mono clip) but a bit noisy)
play(processed_data_qpam, rate=sr)
```

```python
processed_data = quantumaudio.stream(y, shots=8000, scheme = 'qsm') # processing with QSM (quantum state modulation, a LOT slower (>6m for 5s mp3 mono clip) but should sound a bit more accurate (doesn't sound like that though, it's kinda lossy). uses more qubits.)
play(processed_data, rate=sr)
```
